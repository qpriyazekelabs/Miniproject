```python
from datetime import datetime, timedelta
class BikeRental:
    def __init__(self,stock):
        self.bikes = stock
        
    def renthourlybasis(self,id,name,n):
        self.cust_id = id
        self.cust_name = name
        if n > self.bikes:
            return("Bikes are not available")
        else:
            self.bikes = self.bikes - n
            
        
    def hourlybasisbill(self,hourlyrental,n):
        self.hour = hourlyrental
        self.noofbikes = n
        rentalperiod = datetime.now() - self.hour
        if self.noofbikes < 3:
            amount = round(rentalperiod.seconds/3600 * 5) * self.noofbikes
            return amount
        else:
            amount = round (rentalperiod.seconds/3600 * 5) * self.noofbikes
            discount = amount * 0.3
            total = amount -discount
            return total
        
    def rentdailybasis(self,id,name,n):
        self.cust_id = id
        self.cust_name = name
        self.n= n
        if self.n > self.bikes:
            return("Bikes are not available")
        else:
            self.bikes = self.bikes - self.n
            
    def dailybasisbill(self,dailyrental,n):
        self.day = dailyrental
        self.noofbikes = n
        rentaldaily = datetime.now() - self.day
        if self.noofbikes < 3:
            amount = round(rentaldaily.days * 20) * self.noofbikes
            return amount
        else:
            amount = round(rentaldaily.days *20) * self.noofbikes
            discount = amount * 0.3
            total = amount -discount
            return total    
    
    def rentweeklybasis(self,id,name,n):
        self.cust_id = id
        self.cust_name = name
        self.n = n
        if self.n > self.bikes:
            return("Bikes are not availbale")
        else:
            self.bikes = self.bikes - self.n
            
    def weeklybasisbill(self,weeklyrental,n):
            self.week = weeklyrental
            rentweekly = datetime.now() - self.week
            self.noofbikes = n
            if self.noofbikes < 3:
                amount = round(rentweekly.days/7) * 60 * self.noofbikes
                return amount
            else:
                amount = round(rentweekly.days/7) * 60 * self.noofbikes
                discount = amount * 0.3
                total = amount - discount
                return total
    
shop = BikeRental(10)
hourlybasis = shop.renthourlybasis("101","Priya",2)
hour = datetime.now() + timedelta(hours=-1) 
bill = shop.hourlybasisbill(hour,5)
print("****************Bike Rental system**********************")
print("Customer id:")
print(shop.cust_id)
print("Customer Name")
print(shop.cust_name)
print("The amount to be paid in $ is:")
print(bill)
shop1 = BikeRental(20)
dailyrentalbasis = shop1.rentdailybasis("102","Geetha",1)
day = datetime.now() + timedelta(days=-1)
n = 5
bill = shop1.dailybasisbill(day,n)
print("***********************Bike Rental System*************")
print(datetime.now())
print("Customer id:")
print(shop1.cust_id)
print("Customer name")
print(shop1.cust_name)
print("Bike Rented On:")
print(day)
print("No of Bikes Rented")
print(n)
print("The amount to be paid in $ is:")
print(bill)
shop2 = BikeRental(30)
weeklybasis = shop2.rentweeklybasis("104","XYZ",3)
week = datetime.now() + timedelta(weeks=-4)
bill = shop2.weeklybasisbill(week,3)
print("*****************Bike Rental System**********************")
print(datetime.now())
print("Customer id:")
print(shop2.cust_id)
print("Customer name")
print(shop2.cust_name)
print("Bike Rented On")
print(week)
print("The amount to be paid in $ is")
print(bill)




```
